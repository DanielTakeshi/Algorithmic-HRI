```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
figdir = "draft_figures/"
```

```python
actions1 = [20000, 15000, 21000]
actions2 = [49003/3, 13937, 19040, 14317, 1570, 1109]

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,4), gridspec_kw = {'width_ratios':[3, 5]})
title_size = 24
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)

rects1 = ax1.bar(np.arange(len(actions1))+0.125, actions1, width=0.75, color='b')
ax1.set_xticklabels(('NOOP', 'LEFT', 'RIGHT'))
ax1.set_xticks(np.arange(len(actions1))+0.5)
ax1.set_title("Breakout Actions", fontsize=title_size)

rects2 = ax2.bar(np.arange(len(actions2))+0.25, actions2, width=0.75, color='r')
ax2.set_xticklabels(('NOOP', 'FIRE', 'LEFT', 'RIGHT', 'F-L', 'F-R'))
ax2.set_xticks(np.arange(len(actions2))+0.5)
ax2.set_title("Space Invaders Actions", fontsize=title_size)

plt.tight_layout()
plt.savefig(figdir+ "bar_charts_actions.png")
```
